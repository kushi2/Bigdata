```python
def costfn(w,b,daa,da2):
    value1=0
    value2=0
    #c=10
    for i in range(len(daa)):
        val = 1 - da2[i]*(np.dot(w,daa[i]))
        value2 = value2 + max(0,val)
    value1=  0.5*(np.sum(np.square(w))) + c*value2 
    return value1
        
```

```python
def bgd():
    k=0
    li=[]
    fli=[]
    
    delcostis=1
    b=0
    w = np.zeros((1,len(daa[0])))
    f1=costfn(w,b,daa,da2)
    li.append(f1)
    while (delcostis>epsilon1):
        w = w - eta1* funcw(w,b,daa,da2)
        b = b - eta1* funcb(w,b,daa,da2)
        f2=costfn(w,b,daa,da2)
        li.append(f2)
        delcostis=(abs(li[-2]-li[-1])*100)/li[-2]
        fli.append(delcostis)
           
        k=k+1
        print(k)
    return k,fli
```

```python
def funcw(w,b,x,y):
    value1 = 0
    for i in range(len(x)):
        if(y[i]*(np.dot(w,x[i])+b))>= 1:
            value1 += 0
        else:
            value1 += -y[i]*x[i]
    return w+c*value1
```

```python
def funcb(w,b,x,y):
    value1 = 0
    for i in range(len(x)):
        if(y[i]*(np.dot(w,x[i])+b))>= 1:
            value1 += 0
        else:
            value1 += -y[i]
    return c*value1
```

```python
def sgd():
    w2=np.zeros((1,len(x_shuffle[0])))
    b2=0
    l2=0
    k2=0
    batchsize1 = 1
    delcost2 = 0
    coo2 = 1
    li2=[]
    fli2=[]
    n2 = daa.shape[0]
    ff1=costfn(w2,b2,x_shuffle,y_shuffle)
    li2.append(ff1)
    fli2.append(coo2)
    while (coo2>epsilon2):
        
        aa2 = x_shuffle[l2:min(n2,(l2+1))]
        bb2 = y_shuffle[l2:min(n2,(l2+1))]

        w2 = w2 - eta2* funcw(w2,b2,aa2,bb2)
        b2 = b2 - eta2* funcb(w2,b2,aa2,bb2)
        
        
        ff2=costfn(w2,b2,x_shuffle,y_shuffle)
        li2.append(ff2)
        delcostis2=(abs(li2[-2]-li2[-1])*100)/li2[-2]
        coo2 = 0.5*fli2[-1] + 0.5*delcostis2
        fli2.append(coo2)
        l2 = int(l2 % n2) + 1
        k2 = k2+1
        #print(fff2)
        print(k2)
    #print(fli3)
    return k2 , li2
    
    
```
